```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Parameters
k = 0.5      # decreasing hazaard/ infanct mortality
k = 1       # constant hazaard/ random failure scenario
k = 1.4      # insreasing hazaard/ wear out scenario
lam = 20     # scale

t = np.linspace(0.01, 50, 500)

# Weibull Distribution
weibull_pdf = (k / lam) * (t / lam) ** (k - 1) * np.exp(-(t / lam) ** k)

# Generate artificial data based on the Weibull distribution
artificial_data = np.random.weibull(k, 1000) * lam

# Plot histogram of the artificial data and Weibull Distribution on top of each other
plt.figure(figsize=(8, 5))

# Histogram of the artificial data
plt.hist(artificial_data, bins=30, density=True, alpha=0.6, color='blue', label="Histogram")

# Weibull Distribution
plt.plot(t, weibull_pdf, label="Weibull PDF", color='green')

plt.title("Histogram and Weibull Distribution")
plt.xlabel("Time")
plt.ylabel("Density / Probability Density")
plt.legend()

plt.show()
```

```python
# Survival Function
S = np.exp(-(t / lam) ** k)

# Hazard Function
h = (k / lam) * (t / lam) ** (k - 1)


# Plot
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(t, S)
plt.title("Survival Function S(t)")
plt.xlabel("Time")
plt.ylabel("Survival Probability")

plt.subplot(1, 2, 2)
plt.plot(t, h, color='orange')
plt.title("Hazard Function h(t)")
plt.xlabel("Time")
plt.ylabel("Failure Rate")

plt.tight_layout()
plt.show()
```
